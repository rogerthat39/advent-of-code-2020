```python
import math

def getInput(filename):
    tempList = []
    with open(filename, "r") as fp: 
        tempList = [line.strip() for i, line in enumerate(fp)]
    fp.close()
    return tempList

def repeatMap(map):
    #find out how many times to repeat the list - going three across one down you'll need width to be 3x height
    height = len(map)
    width = len(map[0])
    totalRequiredWidth = height*3
    timesToRepeat = math.ceil(totalRequiredWidth / width)
    return [line * timesToRepeat for line in map]

def countTrees(map):
    treeNum = 0
    horizontalPosition = 0 #vertical position will be handled by for loop as always going down 1 each time
    for line in map:
        if line[horizontalPosition] == "#":
            treeNum += 1
        horizontalPosition += 3
    return treeNum

#main routine
lst = repeatMap(getInput("input.txt"))
print(countTrees(lst))
```

```python
import math

def getInput(filename):
    tempList = []
    with open(filename, "r") as fp: 
        tempList = [line.strip() for i, line in enumerate(fp)]
    fp.close()
    return tempList

def repeatMap(map):
    #find out how many times to repeat the list - eg. going three across one down you'll need width to be 3x height
    #now the max width is 7 so will * 7
    height = len(map)
    width = len(map[0])
    totalRequiredWidth = height*7
    timesToRepeat = math.ceil(totalRequiredWidth / width)
    return [line * timesToRepeat for line in map]

def countTrees(map, numRight, numDown):
    treeNum = 0
    horizontalPosition = 0 #vertical position will be handled by for loop
    for i in range(0, len(map), numDown):
        line = map[i]
        if line[horizontalPosition] == "#":
            treeNum += 1
        horizontalPosition += numRight
    return treeNum

#main routine
lst = repeatMap(getInput("input.txt"))
#part 1: print(countTrees(lst, 3))
total = countTrees(lst,1,1) * countTrees(lst,3,1) * countTrees(lst,5,1) * countTrees(lst,7,1) * countTrees(lst,1,2)
print(total)
```
