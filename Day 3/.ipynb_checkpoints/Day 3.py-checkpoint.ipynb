```python
def countTrees(mapOfTrees, numRight, numDown):
    treeNum = 0
    width = len(mapOfTrees[0])
    horizontalPosition = 0 #vertical position will be handled by for loop
    for i in range(0, len(mapOfTrees), numDown):
        line = mapOfTrees[i]
        if line[horizontalPosition%width] == "#": #if horizontal position > width, will loop around back starting from 0
            treeNum += 1
        horizontalPosition += numRight
    return treeNum

#main routine
with open("input.txt", "r") as fp:
    rows = [line.strip() for i, line in enumerate(fp)]
fp.close()
#print(countTrees(rows,3,1))
print(countTrees(rows,1,1) * countTrees(rows,3,1) * countTrees(rows,5,1) * countTrees(rows,7,1) * countTrees(rows,1,2))
```
