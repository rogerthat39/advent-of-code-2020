```python
#part one
#always 128 rows on the plane (0 to 127)
#F and B means Front and Back (where front means lower half of remaining rows)
#there will always be 7 F/B's at the start of the string, then 3 R/L's
#always 8 columns (0 to 7)
#R and L means Right and Left (where left means lower half or remaining columns)

def halveList(num_list, which_half_to_keep):
    #print(num_list)
    if which_half_to_keep == 0: #keep first/lower half
        return num_list[:int(len(num_list)/2)]
    elif which_half_to_keep == 1: #keep latter/higher half
        return num_list[int(len(num_list)/2):]

#initial code for part one
def getSeatIDs():
    seat_ids = []
    for boarding_pass in boarding_passes:
        possible_rows = [num for num in range(0,128)]
        possible_columns = [num for num in range(0,8)]

        for letter in boarding_pass[:7]:
            possible_rows = halveList(possible_rows, 0) if letter == "F" else halveList(possible_rows, 1)
            #print(f"{possible_rows[0]} to {possible_rows[-1]}")

        for letter in boarding_pass[-3:]:
            #print(letter)
            possible_columns = halveList(possible_columns, 0) if letter == "L" else halveList(possible_columns, 1)
            #print(f"{possible_columns[0]} to {possible_columns[-1]}")

        seat_ids.append(possible_rows[0] * 8 + possible_columns[0])
        #print(seat_ids)
    return seat_ids


#main routine
with open("input.txt", "r") as fp:
    boarding_passes = [line.strip() for i, line in enumerate(fp)]
fp.close()

seat_ids = sorted(getSeatIDs())
min_id = seat_ids[0]
max_id = seat_ids[-1]

#part one answer:
print(f"Part one: {max_id}")
   
#part two
#your boarding pass is the only one missing from this list
#and will be between the max and min numbers
for seat_id in range(min_id, max_id):
    if seat_id not in seat_ids:
        print(f"Part two: {seat_id}")
        break
```

```python
#add tests
def test():
    boarding_info = []
    test_data = ["FBFBBFFRLR","BFFFBBFRRR","FFFBBBFRRR","BBFFBBFRLL","FFFFFFBLLR","FFFFFFFLLL","BBBBBBBRRR"]
    for test_string in test_data:
        boarding_pass = test_string.strip().replace("F", "0").replace("L", "0").replace("B", "1").replace("R", "1")
        row = int(boarding_pass[:7], 2)
        column = int(boarding_pass[-3:], 2)
        seat_id = int(boarding_pass, 2)
        boarding_info.append((row, column, seat_id))
       
        #row * 8 + column is the same as 2^3 which is the same as shifting 'row' 3 to the left
        #ie. treating the whole number as binary
        assert row * 8 + column == seat_id

    #check row, column and seat_id are correct
    #in format:(row, column, seat_id)
    assert boarding_info[0] == (44, 5, 357)
    assert boarding_info[1] == (70, 7, 567)
    assert boarding_info[2] == (14, 7, 119)
    assert boarding_info[3] == (102, 4, 820)
    assert boarding_info[4] == (1, 1, 9)
    assert boarding_info[5] == (0, 0, 0)
    assert boarding_info[6] == (127, 7, 1023)

#run tests
test()


def calculateSeatId(boarding_pass):
    #where it says "take lower half" replace with 0, "take higher half" replace with 1
    boarding_pass = boarding_pass.replace("F", "0").replace("L", "0").replace("B", "1").replace("R", "1")
    #return an integer (,2 treats the string as binary)    
    return int(boarding_pass, 2)

#main routine
seat_ids = []
with open("input.txt", "r") as fp:
    for i, line in enumerate(fp):
        seat_ids.append(calculateSeatId(line.strip()))
fp.close()

#sort the list so we can easily get the min and max values
seat_ids = sorted(seat_ids)
min_id = seat_ids[0]
max_id = seat_ids[-1]

#part one - get maximum seat id
print(f"Part one: {max_id}")

#part two - get missing seat id
#your seat will not be on either end ie. has to be between min and max
for potential_id in range(min_id, max_id):
    if potential_id not in seat_ids:
        print(f"Part two: {potential_id}")
        break
```
