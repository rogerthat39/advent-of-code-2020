```python
#part 1
def getInput(filename):
    tempList = []
    with open(filename, "r") as fp: 
        for i, line in enumerate(fp):
            tempLine = line.split(" ")
            tempList.append({
                "lowerNum": int(tempLine[0].split("-")[0]),
                "upperNum": int(tempLine[0].split("-")[1]),
                "letter":   tempLine[1][0],
                "password": tempLine[2].rstrip('\n')
            })
    fp.close()
    return tempList
    
def checkPasswordValidity(inp):
    count = 0
    for dict in inp:
        letterCount = dict["password"].count(dict["letter"])
        #count number of letter in string
        #see if that number is between num1 and num2, if so, add 1 to count
        if letterCount >= dict["lowerNum"] and letterCount <= dict["upperNum"]:
            count += 1
    return count

#main routine
passwords = getInput("input.txt")
print(f"Number of valid passwords: {checkPasswordValidity(passwords)}")
```

```python
def getInput(filename):
    tempList = []
    with open(filename, "r") as fp: 
        for i, line in enumerate(fp):
            tempLine = line.split(" ")
            tempList.append({
                "lowerNum": int(tempLine[0].split("-")[0]),
                "upperNum": int(tempLine[0].split("-")[1]),
                "letter":   tempLine[1][0],
                "password": tempLine[2].rstrip('\n')
            })
    fp.close()
    return tempList
    
def checkPasswordValidity1(inp):
    count = 0
    for dict in inp:
        #count number of letter in string
        letterCount = dict["password"].count(dict["letter"])
        #see if that number is between num1 and num2, if so, add 1 to count
        if letterCount >= dict["lowerNum"] and letterCount <= dict["upperNum"]:
            count += 1
    return count

def checkPasswordValidity2(inp):
    count = 0
    for dict in inp:
        firstLetter = dict["password"][dict["lowerNum"]-1]
        secondLetter = dict["password"][dict["upperNum"]-1]
        #one of either the first or last letter should match, the other should not match
        if firstLetter == dict["letter"] and secondLetter != dict["letter"]:
            count += 1
        elif firstLetter != dict["letter"] and secondLetter == dict["letter"]:
            count += 1
    return count

#main routine
passwords = getInput("input.txt")
print(f"Number of valid passwords (1): {checkPasswordValidity1(passwords)}")
print(f"Number of valid passwords (2): {checkPasswordValidity2(passwords)}")
```
